# Import the datasets for steps

```r
library(readr)
library(dplyr)
# Import datasets: 3/4 tables (daily, hourly, Minutes wide and narrow)
dailySteps_merged <- read_csv("/kaggle/input/bellabeat/Fitabase Data 4.12.16-5.12.16/dailySteps_merged.csv")
hourlySteps_merged <- read_csv("/kaggle/input/bellabeat/Fitabase Data 4.12.16-5.12.16/hourlySteps_merged.csv")
minuteStepsNarrow_merged <- read_csv("/kaggle/input/bellabeat/Fitabase Data 4.12.16-5.12.16/minuteStepsNarrow_merged.csv")
```

## Group, sort and ﬁlter the data

```r
hourlySteps_merged <- hourlySteps_merged %>% group_by(Id)
hourlySteps_merged

minuteStepsNarrow_merged <- minuteStepsNarrow_merged %>% group_by(Id)
minuteStepsNarrow_merged

dailySteps_merged <- dailySteps_merged %>% group_by(Id)
dailySteps_merged
```

## Check for null rows in dailySteps_merged

```r
is.null(dailySteps_merged$Id)
is.null(dailySteps_merged$StepTotal)
is.null(dailySteps_merged$ActivityDay)
```

## Check for null rows in hourlySteps_merged

```r
is.null(hourlySteps_merged$Id)
is.null(hourlySteps_merged$StepTotal)
is.null(hourlySteps_merged$ActivityHour)
```

## Check for null rows in minuteStepsNarrow_merged

```r
is.null(minuteStepsNarrow_merged$Id)
is.null(minuteStepsNarrow_merged$Steps)
is.null(minuteStepsNarrow_merged$ActivityMinute)
```
